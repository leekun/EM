# 总结。Markov Chain Monte Carlo (MCMC)

参考1：https://cosx.org/2013/01/lda-math-mcmc-and-gibbs-sampling

** 总结 **：

    -MCMC 的目的:  对于概率分布p(x)，如何生成它的样本。
    -MCMC 的优势:  MCMC定义了一种框架，对任意的概率分布，都可以进行模拟，进行抽样。
    -MCMC 的注意点: MCMC是关联采样，样本间不是独立的。即样本和样本间存在"转移概率"
    -MCMC及其改进算法Metropolis-Hastings，gibbs采样，都是在构造转移矩阵。
    
** 其他 **：
  
    -第二个MC是指：随机模拟 (或者统计模拟) 别名是蒙特卡罗方法(Monte Carlo Simulation)


# 简单的抽样的方法：

## 常见分布的抽样：
    - 均匀分布 U(0,1), 线性同余发生器
    - 常见概率分布，无论连续的，还是离散的分布，都可以基于U(0,1)样本来生成。
    - 如正太分布，可通过BOX-Muller变换:  
$$U_1, U_2 \sim Uniform[0,1]$$ 
$$Z_0  = \sqrt{-2\ln U_1} cos(2\pi U_2)$$ 
$$Z_1  = \sqrt{-2\ln U_1} sin(2\pi U_2)$$
    - 其他的分布如：指数分布、Gamma 分布、t 分布、F 分布、Beta分布、Dirichlet分布等，也都可以通过类似的数学变换得到
    - 这些采样法法，样本之间是独立的。
    
## 拒绝接受采样（Acceptance-Rejection Sampling）
1, 需求： 根某已知分布的概率密度函数 f(x) ，产生服从此分布的样本 X

2, 准备工作：

&nbsp;&nbsp;&nbsp;&nbsp;2.1 需要一个辅助的“建议分布proposal distribution” G （已知其概率密度函数 g(y) ）来产生候选样本。——由分布来产生候选样本，因此需要我们能够从 G 抽样。即我们必须能够生成服从此概率分布的样本 Y 。比如可以选择均匀分布、正态分布。

&nbsp;&nbsp;&nbsp;&nbsp;2.2 还需要另一个辅助的均匀分布 U(0,1) 

&nbsp;&nbsp;&nbsp;&nbsp;2.3 计算一个常数值 c 。——满足不等式 $ c∗g(x)\geqslant f(x) $ 的最小值 c （当然，我们非常希望 c 接近于1)

3, 开始样本生成：

&nbsp;&nbsp;&nbsp;&nbsp; 3.1 从建议分布 G 抽样，得到样本 Y 。

&nbsp;&nbsp;&nbsp;&nbsp; 3.2 从分布 U(0,1) 抽样，得到样本 U 。

&nbsp;&nbsp;&nbsp;&nbsp; 3.3如果 $ U\leqslant\frac{f(Y)}{c*g(Y)} $，则令 X=Y （接受 Y ），否则继续执行步骤1（拒绝）。

 
     
# MCMC采样概述：

## 马氏链定理：

如果有一个非周期的马氏链，其转移概率矩阵为P, 且它的任意两个状态是联通的，那么
         
$ \lim_{n\rightarrow\infty}P_{ij}^n $存在，且与i无关，记$ \lim_{n\rightarrow\infty}P_{ij}^n = \pi(j) $. 我们有：
    
1，$ \lim_{n \rightarrow \infty} P^n =\begin{bmatrix}
\pi(1) ,  \pi(2) \cdots \pi(j) \cdots \\
\pi(1) , \pi(2) \cdots  \pi(j)  \cdots \\
\cdots  \cdots  \cdots  \cdots  \cdots \\  
\pi(1) , \pi(2)  \cdots  \pi(j)  \cdots \\ 
\cdots  \cdots  \cdots  \cdots  \cdots \\ 
\end{bmatrix} $  

2，$ \pi(j) = \sum_{i=0}^{\infty}\pi(i)P_{ij} $ 

3，$\pi$ 是方程 $\pi P = \pi$的唯一非负解。 其中
    $$ \pi = [\pi(1), \pi(2), \cdots, \pi(j),\cdots ], \quad \sum_{i=0}^{\infty} \pi_i = 1 $$
    


## 根据上面的定理，我们可以得到MCMC的基本步骤：

1，定义1：$ \pi(x) $ 为我们需要采样的分布。$ x_1、x_2 、x_3  \cdots $  为所有满足分布$ \pi(x) $ 的样本点（概率为连续分布时,无限个点), 他们对应的概率为$ \pi(x_1)、\pi(x_2) 、\pi(x_3)  \cdots $  
    
2，定义2：这些样本点以转移概率矩阵P可以相互的”转移”, P 已知【或可构造出P, 具体构造步骤即为MCMC和gibbs采样算法核心】。 $ \pi(x) $ 为 P 的平稳分布。 

3，定义3：对于初始的任意一个分布 $ \pi_0(x) $，它在$ x_1、x_2 、x_3  \cdots $ 上的概率为$ \pi_0(x_1)、\pi_0(x_2) 、\pi_0(x_3)  \cdots $，由于是概率分布，其和为1 。 定义随机变量 $ X_0 \sim \pi_0(x) $
    
4，具体步骤：

   转移：通过P转移一次后变为分布$ \pi_1(x) $，它在$ x_1、x_2 、x_3  \cdots $ 上的概率为$ \pi_1(x_1)、\pi_1(x_2) 、\pi_1(x_3)  \cdots $。 定义随机变量 $ X_1 \sim \pi_1(x) $
   
   收敛：以此不断的转移，当t步时 $ P^t $ 收敛，那么 $ \pi_t(x) $ 即为 $ \pi(x) $ 分布 , $ X_t ~ \sim \pi_t(x) $
   
   采样：按照 $ X_t $ 在各点的概率（分布）情况( $ \pi_t(x_1)、\pi_t(x_2) 、\pi_t(x_3)  \cdots $ )，选取 $ x_t $ 作为 此步的样本。 【此步采样有技巧，具体看mcmc和gibbs算法； 否认又成了要解决的根据分布采样问题了】
   
   采样2：在$ x_t $的节点上，按照马尔科夫链的转移矩阵P再转移一次后的变量分布仍然满足  $ X_{t+1} ~ \sim \pi_{t+1}(x) $ ，再次采样$ x_{t+1} $。
   
   ** 不断的重复上面的步骤 **



##  物理意义：

   $ \pi_t(x) P $ 就相当于所有的样本点在马尔科夫链上转移一次后得到的新的分布。
   
   由分布$ \pi_t(x) $ 采一个样得：$ x_t $ 。在$ x_t $ 这个点上，根据转移矩阵，再转移一次，采样得到的 $ x_{t+1} $ 也会满足分布$ \pi_t(x) P $


# MCMC 算法：

根据上面的讨论，我们知道只要可以构造出转移矩阵P 使得 $ \pi_(x) $ 为其平稳分布即可。 问题化为如何构造P

**细致平稳条件：**

   若满足 $ \pi(i)P_{ij} = \pi(j)P_{ji} \quad\quad \text{for all} \quad i,j $ 则 $ \pi_(x) $是马氏链的平稳分布 

那么问题转化为，如何构造P使得其满足细致平稳条件？


## 构造转移矩阵

1，构造**任意**的转移矩阵Q，q(i,j)表示第i个样本，转移到第j个样本的概率

2，对于**任意**的转移矩阵Q来说，通常 $ \pi(i) q(i,j) \neq \pi(j) q(j,i)$ 细致条件不满足。
如何改造：


$$ \alpha(i,j)= \pi(j) q(j,i)， \quad \alpha(j,i) = \pi(i) q(i,j) $$

下式就成立了：

$$ \pi(i) \underbrace{q(i,j)\alpha(i,j)}_{Q'(i,j)} = \pi(j)\underbrace{q(j,i)\alpha(j,i)}_{Q'(j,i)} \quad (**) $$

构造出的$Q’$，即为所求
也即：
$$ \pi(i) \underbrace{q(i,j) \pi(j) q(j,i)}_{Q'(i,j)} = \pi(j)\underbrace{q(j,i) \pi(i) q(i,j) }_{Q'(j,i)} \quad (**) $$


## 采样步骤-采第t+1个样本：** （类似于拒绝接受采样）
1, 先根据 转移矩阵Q的一行 $ q(x_t, y)$ 采样得到y 。 由于通常定义转移矩阵Q的一行是常见分布,如正太分布，则此采样很容易实现。

2, 均匀分布采样 $ U(0,1) $ 采样得到 $u$

3, 若 $ u < \alpha(x_t, y)= \pi(y) q(y,x_t) $ 则接受转移 $x_{t+1} = y $, 并放入样本池中。

4, 否则不接受转移, 重新走第一步, 不生成新的样本


## 细节：
1, 初始Q任意的意思就是说，需要Q的每一行加起来为1就可以了，Q的每一行都是个条件分布。 但是注意每行不能一样，否则有问题，请见5.2

&nbsp;&nbsp;&nbsp;&nbsp; 1.1 如当 $ \pi(x) $为连续分布，可以定义Q每一行为一个正太分布,均值为上个采样点。
    
&nbsp;&nbsp;&nbsp;&nbsp; 1.2 如当 $ \pi(x) $为离散分布，可以定义各行为均匀分布。

2, 样本可以是一维的，也可以是高维的

3, 采样步骤，是先选择一个候选的，然后判断是否满足第三步的条件。若是第三步的概率很小，那么收敛的速度也太慢。可以改进得到Metropolis-Hastings 算法


# Metropolis-Hastings 算法实例

```python
# 参考1： http://www.cnblogs.com/pinard/p/6638955.html
# 正太分布(均值3，方差2)抽样

import random
import math
from scipy.stats import norm
import matplotlib.pyplot as plt
%matplotlib inline

def norm_dist_prob(theta):
    y = norm.pdf(theta, loc=3, scale=2)
    return y

sigma = 1

def trans_prob(mean,x):
    y = norm.pdf(x, loc=mean, scale=sigma)
    return y

T = 5000
pi = [0]*T
t = 0
while t < T-1:
    t = t + 1

    # Q的一行为正太分布：均值为上一个样本值。标准差为sigma。size个变量。
    y = norm.rvs(loc=pi[t - 1], scale=sigma, size=1, random_state=None)
    
    p_xt = norm_dist_prob(pi[t - 1])
    p_y = norm_dist_prob(y[0])
    
    q_y2xt= trans_prob(y[0],pi[t - 1])
    q_xt2y = trans_prob(pi[t - 1],y[0])
    
    alpha = min(1, p_y*q_y2xt/p_xt/q_xt2y)
    
    u = random.uniform(0, 1)
    if u < alpha:
        pi[t] = y[0]
    else: 
        t = t-1
        # 此时拒绝，就不能放到样本集合中了, 需要继续迭代x_t+1。
        #pi[t] = pi[t - 1]

# 假设100步后收敛，只画100步之后的样本
plt.scatter(pi[100:], norm.pdf(pi[100:], loc=3, scale=2)) 
num_bins = 50
plt.hist(pi[100:], num_bins, normed=1, facecolor='red', alpha=0.7)
plt.show()
```

## 转移矩阵 不可任意定义

```python
# 参考1： http://www.cnblogs.com/pinard/p/6638955.html
# 正太分布(均值3，方差2)抽样

import random
import math
from scipy.stats import norm
import matplotlib.pyplot as plt
%matplotlib inline

def norm_dist_prob(theta):
    y = norm.pdf(theta, loc=3, scale=2)
    return y

sigma = 1

def trans_prob(mean,x):
    y = norm.pdf(x, loc=0, scale=sigma) 
    # 注：当每行都选择同样的转移概率的时候，采样结果，不再是需要的分布(均值3，方差2)

    return y

T = 5000
pi = [0]*T
t = 0
while t < T-1:
    t = t + 1

    # Q的一行为正太分布：均值为上一个样本值。标准差为sigma。size个变量。
    y = norm.rvs(loc=0, scale=sigma, size=1, random_state=None)
    # 注：当每行都选择同样的转移概率的时候，采样结果，不再是需要的分布(均值3，方差2)

    
    p_xt = norm_dist_prob(pi[t - 1])
    p_y = norm_dist_prob(y[0])
    
    q_y2xt= trans_prob(y[0],pi[t - 1])
    q_xt2y = trans_prob(pi[t - 1],y[0])
    
    alpha = min(1, p_y*q_y2xt/p_xt/q_xt2y)
    

    u = random.uniform(0, 1)
    if u < alpha:
        pi[t] = y[0]
    else: 
        t = t-1
        # 此时拒绝，就不能放到样本集合中了, 需要继续迭代x_t+1。
        #pi[t] = pi[t - 1]

# 假设100步后收敛，只画100步之后的样本
plt.scatter(pi[100:], norm.pdf(pi[100:], loc=3, scale=2)) 
num_bins = 50
plt.hist(pi[100:], num_bins, normed=1, facecolor='red', alpha=0.7)
plt.show()
```
